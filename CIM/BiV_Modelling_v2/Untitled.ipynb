```python
import os
import sys

import numpy as np
import pandas as pd
```

```python
for patient in os.listdir('./test_data/'):
    print(patient)
    
    # load GP_ED.txt
    gp_ed = pd.read_csv(os.path.join('./test_data/', patient, 'GP_ES.txt'), delimiter='\t')
    # find septal points
    lax_septum = gp_ed[gp_ed['contour type'] == 'LAX_RV_SEPTUM']
    points = [[x['x'], x['y'], x['z']] for i, x in lax_septum.iterrows()]
    
    # find freewall points
    lax_freewall = gp_ed[gp_ed['contour type'] == 'LAX_RV_FREEWALL']
    for i, row in lax_freewall.iterrows():
        point = [row['x'], row['y'], row['z']]

        # remove duplicate poitns
        if point in points:
            gp_ed = gp_ed.drop(index=i)
            print(i)
    
    gp_ed.to_csv(os.path.join('./test_data/', patient, 'GP_ES.txt'), sep = '\t', index=False)
```
